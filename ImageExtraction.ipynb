```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread('/content/leopard.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
# print image shape
print(image.shape)
# print gray scale shape
print(gray.shape)
```

```python
# extract R G B from image
R,G,B=image[:,:,0],image[:,:,1],image[:,:,2]
# create blank images for visualization
zeros=np.zeros_like(R)
# convert single channels into three color channels
red=np.stack([R,zeros,zeros],axis=-1)
green=np.stack([zeros,G,zeros],axis=-1)
blue=np.stack([zeros,zeros,B],axis=-1)
# display image
plt.imshow(red)
plt.show()
plt.imshow(green)
plt.show()
plt.imshow(blue)
plt.show()
```
